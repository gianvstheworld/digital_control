Para isso, considere o seguinte exemplo:
$$
G(s) = \frac{s + 3}{(s + 1)(s + 2)}
$$

A função de transferência equivalente em tempo discreto pode ser computada usando o Matlab através da função `c2d` para os métodos degrau-invariante (zoh), impulso-invariante (impulse), mapeamento casado de polos e zeros (matched), método Bilinear (tustin) além de outros. (Dica: consulte a documentação do Matlab sobre discretização) https://www.mathworks.com/help/control/ug/continuous-discrete-conversion-methods.html

Por exemplo, o código a seguir obtém a FT em tempo discreto usando o método degrau-invariante, impulso-invariante, mapeamento casado de polos e zeros e método de tustin:

1. `close all` % fecha todas janelas  
2. `clear all` % limpa memória  
3. `clc` % limpa command window  
4. `%%`  
5. `num = [1 3];`  
6. `den = conv([1 1], [1 2]);`  
7. `G = tf(num, den);` % FT em tempo contínuo  
8. `T = 0.1;` % tempo de amostragem  
9. `Gd1 = c2d(G, T, 'zoh');` % obtém a FT discreta usando método degrau-invariante  
10. `Gd2 = c2d(G, T, 'impulse');` % obtém a FT discreta usando método impulso-invariante  
11. `Gd3 = c2d(G, T, 'matched');` % obtém a FT discreta usando método mapeamento casado de polos/zeros

O que pode-se concluir através da análise da resposta em frequência para os diferentes métodos de discretização?


```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
import scipy.signal as signal
import control

from scipy.signal import lti, freqz, bode, cont2discrete
```

```python
def create_discrete_systems(num, den, T):
    # Convertendo para o domínio discreto usando diferentes métodos
    Gd1 = cont2discrete((num, den), T, method='zoh')
    num_d1, den_d1, _ = Gd1

    Gd2 = cont2discrete((num, den), T, method='impulse')
    num_d2, den_d2, _ = Gd2

    Gd4 = cont2discrete((num, den), T, method='tustin')
    num_d4, den_d4, _ = Gd4

    # Discrete systems (ZOH, Impulse, Tustin)
    discrete_systems = [(num_d1, den_d1, 'ZOH'),
                        (num_d2, den_d2, 'Impulse'),
                        (num_d4, den_d4, 'Tustin')]
    
    return discrete_systems
```

```python
def plot_bode(num, den, discrete_systems, T):
    # Frequency range for continuous system (from 10^-4 to 10^2 rad/s)
    w_cont = np.logspace(-2, 2, 8000)  

    # Continuous system
    G_cont = lti(num, den)  # Create continuous system
    w_cont, mag_cont, phase_cont = bode(G_cont, w=w_cont)  # Get Bode plot for continuous system

    colors = ['red', 'green', 'orange']  # Colors for different discrete systems

    # Loop through each discrete system and pair it with the continuous system
    for i, (num_d, den_d, title) in enumerate(discrete_systems):
        num_d = num_d.flatten()  # Ensure num_d is 1D
        den_d = den_d.flatten()  # Ensure den_d is 1D

        # Frequency response for the discrete system
        w_discrete, h_discrete = freqz(num_d, den_d, worN=8000)

        # Convert discrete frequency (in normalized units) to continuous frequency (rad/s)
        w_discrete = w_discrete * (1 / T)

        # Filter frequencies to only plot from 10^-0 onwards
        mask_cont = w_cont >= 10**-0
        mask_discrete = w_discrete >= 10**-0

        # Create a new figure for each pair of continuous and discrete systems
        plt.figure()

        # Magnitude plot: continuous (blue) and discrete (paired color)
        plt.subplot(2, 1, 1)
        plt.semilogx(w_cont[mask_cont], mag_cont[mask_cont], label='Continuous System', color='blue')  # Continuous system
        plt.semilogx(w_discrete[mask_discrete], 20 * np.log10(np.abs(h_discrete[mask_discrete])), 
                     label=f'{title} (Discrete)', color=colors[i])  # Discrete system
        plt.title(f'Bode Plot: Continuous vs {title}')
        plt.ylabel('Magnitude [dB]')
        plt.grid(True)
        plt.legend()

        # Phase plot: continuous (blue) and discrete (paired color)
        plt.subplot(2, 1, 2)
        plt.semilogx(w_cont[mask_cont], phase_cont[mask_cont], label='Continuous System', color='blue')  # Continuous system
        plt.semilogx(w_discrete[mask_discrete], np.angle(h_discrete[mask_discrete], deg=True), 
                     label=f'{title} (Discrete)', color=colors[i])  # Discrete system
        plt.ylabel('Phase [degrees]')
        plt.xlabel('Frequency [rad/s]')
        plt.grid(True)
        plt.legend()

    plt.show()
```

```python
# Tempo de amostragem
T = 0.1

a = 1
num1 = np.array([1, 0])
den1 = np.convolve([1, a], [1, a])

num2 = np.array([1, 3])
den2 = np.convolve([1, 1], [1, 2])

num3 = np.array([1, 0])
den3 = np.array([1, 0, 3, 2, 1])

ds1 = create_discrete_systems(num1, den1, T)
ds2 = create_discrete_systems(num2, den2, T)
ds3 = create_discrete_systems(num3, den3, T)

print('-'*100)
plot_bode(num1, den1, ds1, T)
print('-'*100)
plot_bode(num2, den2, ds2, T)
print('-'*100)
plot_bode(num3, den3, ds3, T)
```

### Questão 2

Nessa atividade, o objetivo é usar o recurso de cálculo simbólico do Matlab para obter as FT equivalentes em tempo discreto. Para isso, considere o seguinte exemplo:

$$
G(s) = \frac{a}{s + a}
$$

Para o caso do método de Euler-Backward de discretização, pode-se implementar conforme o código de exemplo a seguir:

1. `close all` % fecha todas janelas  
2. `clear all` % limpa memória  
3. `clc` % limpa command window  
4. `%%`  
5. `syms s z T` % define variáveis simbólicas  
6. `G = a / (s + a);` % monta FT em tempo contínuo  
7. `Gd = subs(G, 's', (z - 1) / (z * T));` % obtém FT discreta via substituição  
8. `pretty(Gd);` % mostra FT de forma "bonita"  

Seguindo a mesma lógica, encontre a FT em tempo discreto usando os métodos de Euler-Forward e Euler-Backward dos seguintes casos:

1. $$ G(s) = \frac{1}{s^2} $$  
2. $$ G(s) = \frac{s + 2}{(s + 1)(s + 5)} $$  
3. $$ G(s) = \frac{s + 3}{s^2 (s + 1)(s - 9)} $$  


```python
s, z, T = sp.symbols('s z T')
a = sp.symbols('a')

# Etapa 3: Definir as funções de transferência contínuas G(s)
G1 = 1 / s**2
G2 = (s + 2) / ((s + 1) * (s + 5))
G3 = (s + 3) / (s**2 * (s + 1) * (s - 9))

# Etapa 4: Euler-Backward - Substituir s por (z - 1) / (z * T)
Gd1_backward = G1.subs(s, (z - 1) / (z * T))
Gd2_backward = G2.subs(s, (z - 1) / (z * T))
Gd3_backward = G3.subs(s, (z - 1) / (z * T))

# Etapa 5: Euler-Forward - Substituir s por (z - 1) / T
Gd1_forward = G1.subs(s, (z - 1) / T)
Gd2_forward = G2.subs(s, (z - 1) / T)
Gd3_forward = G3.subs(s, (z - 1) / T)

# Etapa 6: Exibir as funções de transferência discretas de forma "bonita"
print("Euler-Backward Method:")
sp.pretty_print(Gd1_backward)
sp.pretty_print(Gd2_backward)
sp.pretty_print(Gd3_backward)

print("\nEuler-Forward Method:")
sp.pretty_print(Gd1_forward)
sp.pretty_print(Gd2_forward)
sp.pretty_print(Gd3_forward)
```

Seguindo a mesma lógica, compare os métodos de discretização degrau-invariante e impulso-invariante para a seguinte função de transferência com $T = 0.1s$ e $t \in [0, 5]$:

$$
G(s) = \frac{(s + 2)(s + 1)}{(s + 5)(s + 3)(s + 7)}
$$

O que pode-se observar em relação às respostas temporais dos dois métodos de discretização para as entradas em degrau e impulso?


```python
def define_responses(num, den):

    # Criar a função de transferência contínua
    G_s = signal.TransferFunction(num.coeffs, den.coeffs)

    # Etapa 2: Definir tempo de amostragem e intervalo de tempo contínuo
    T = 0.2  # Tempo de amostragem
    t_continuo = np.arange(0, 5, 0.001)  # Vetor de tempo contínuo

    # Etapa 3: Obter a resposta ao degrau e ao impulso em tempo contínuo
    t_step_continuo, y_step_continuo = signal.step(G_s, T=t_continuo)
    t_impulse_continuo, y_impulse_continuo = signal.impulse(G_s, T=t_continuo)

    # Etapa 4: Discretizar a função de transferência
    Gd_step_invariant = signal.cont2discrete((num.coeffs, den.coeffs), T, method='zoh')  # degrau-invariante
    Gd_impulse_invariant = signal.cont2discrete((num.coeffs, den.coeffs), T, method='impulse')  # impulso-invariante

    # Etapa 5: Definir intervalo de tempo discreto
    t_discreto = np.arange(0, 5, T)

    # Etapa 6: Obter resposta ao degrau e ao impulso em tempo discreto
    _, y_step_discrete_zoh = signal.dstep(Gd_step_invariant, t=t_discreto)
    _, y_impulse_discrete_zoh = signal.dimpulse(Gd_step_invariant, t=t_discreto)

    _, y_step_discrete_impulse = signal.dstep(Gd_impulse_invariant, t=t_discreto)
    _, y_impulse_discrete_impulse = signal.dimpulse(Gd_impulse_invariant, t=t_discreto)

    # Ajustar as saídas para ficarem no formato correto
    y_step_discrete_zoh = np.squeeze(y_step_discrete_zoh)
    y_impulse_discrete_zoh = np.squeeze(y_impulse_discrete_zoh)
    y_step_discrete_impulse = np.squeeze(y_step_discrete_impulse)
    y_impulse_discrete_impulse = np.squeeze(y_impulse_discrete_impulse)

    discrete_responses = [(t_discreto, y_step_discrete_zoh, y_impulse_discrete_zoh, 'ZOH'),
                            (t_discreto, y_step_discrete_impulse, y_impulse_discrete_impulse, 'Impulse')]
    
    continous_responses = [(t_continuo, y_step_continuo, t_impulse_continuo, y_impulse_continuo)]

    return continous_responses, discrete_responses
```

```python
def plot_responses(continous_responses, discrete_responses):

    # Descompactar as respostas contínuas
    t_continuo, y_step_continuo, t_impulse_continuo, y_impulse_continuo = continous_responses[0]

    # Descompactar as respostas discretas
    t_discreto, y_step_discrete_zoh, y_impulse_discrete_zoh, title_zoh = discrete_responses[0]
    _, y_step_discrete_impulse, y_impulse_discrete_impulse, title_impulse = discrete_responses[1]

    # Plot 1: Resposta ao degrau
    plt.figure()
    plt.plot(t_continuo, y_step_continuo, label='Contínuo (Degrau)', linewidth=1.5)
    plt.step(t_discreto, y_step_discrete_zoh, label='Degrau-Invariante', where='post', linewidth=1.5)
    plt.step(t_discreto, y_step_discrete_impulse, label='Impulso-Invariante', where='post', linewidth=1.5)
    plt.title('Resposta ao Degrau')
    plt.xlabel('Tempo [s]')
    plt.ylabel('Amplitude')
    plt.legend()
    plt.grid(True)

    # Plot 2: Resposta ao impulso
    plt.figure()
    plt.plot(t_continuo, y_impulse_continuo, label='Contínuo (Impulso)', linewidth=1.5)
    plt.step(t_discreto, 5*y_impulse_discrete_zoh, label='Degrau-Invariante', where='post', linewidth=1.5)
    plt.step(t_discreto, 5*y_impulse_discrete_impulse, label='Impulso-Invariante', where='post', linewidth=1.5)
    plt.title('Resposta ao Impulso')
    plt.xlabel('Tempo [s]')
    plt.ylabel('Amplitude')
    plt.legend()
    plt.grid(True)

    # Exibir os gráficos
    plt.show()
```

```python
# Etapa 1: Definir a função de transferência contínua G(s)
num = np.poly1d([1, 3])
den = np.poly1d([1, 1, 1])

# Etapa 2: Definir as respostas contínuas e discretas
continous_responses, discrete_responses = define_responses(num, den)

# Etapa 3: Exibir as respostas contínuas e discretas
plot_responses(continous_responses, discrete_responses)

print('-'*100)
#-------- exercicio 

num = np.poly1d(np.convolve([1, 2], [1, 1]))  # (s + 2)(s + 1)
den = np.poly1d(np.convolve(np.convolve([1, 5], [1, 3]), [1, 7]))  # (s + 5)(s + 3)(s + 7)

continous_responses, discrete_responses = define_responses(num, den)
plot_responses(continous_responses, discrete_responses)
```

Observa-se que ambos os métodos não podem ser usados invariavelmente para discretizar sistemas contínuos. O método degrau-invariante é mais adequado para sistemas de primeira ordem, enquanto o método impulso-invariante é mais adequado para sistemas de segunda ordem.

Para isso, considere uma equação de primeira ordem na forma:

$$
\frac{dy}{dt} + y = u(t)
$$

A função de transferência associada a essa EDO é:

$$
G(s) = \frac{1}{s + 1}
$$

Pode-se obter a FT equivalente em tempo discreto usando o método de degrau-invariante no Matlab através do seguinte código:

O que resulta em:

$$
G_d(z) = \frac{0.09516}{z - 0.9048} = \frac{0.09516z^{-1}}{1 - 0.9048z^{-1}}
$$

Portanto, como $G_d(z) = \frac{Y_d(z)}{U_d(z)}$, tem-se a equação a diferenças associada dada por:

$$
y[k] = 0.9048y[k - 1] + 0.09516u[k - 1]

3. `u = ones(size(td));`

```python
# Número de amostras para o sistema discreto
num_samples = 500
t_continuo = np.linspace(0, num_samples * T, num_samples)

# Resposta ao degrau do sistema contínuo
t_c, y_c = control.step_response(G, T=t_continuo)

# Resposta ao degrau do sistema discreto
num_d = Gd.num[0][0]
den_d = Gd.den[0][0]
sys_d = signal.dlti(num_d, den_d, dt=T)

t_d, y_d = signal.dstep(sys_d, n=num_samples)
y_d = np.squeeze(y_d)  # Remover dimensões desnecessárias
t_d = np.arange(0, T * num_samples, T)  # Criar vetor de tempo discreto

# Plotar Resposta ao Degrau
plt.figure(figsize=(10, 6))
plt.plot(t_c, y_c, label='Contínuo', linewidth=3)
plt.step(t_d, y_d, linestyle='--', where='post', label='Discreto', linewidth=2)
plt.xlabel('Tempo (s)')
plt.ylabel('Amplitude')
plt.title('Resposta ao Degrau: Contínuo vs Discreto')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Extração dos coeficientes da função de transferência discreta
num_d = Gd.num[0][0]
den_d = Gd.den[0][0]

# Coeficientes da função de transferência discreta
a1 = den_d[1]
b0 = num_d[0]
b1 = num_d[1]

print("\nEquação a Diferenças:")
print(f"y[k] = {a1} * y[k-1] + {b0} * u[k] + {b1} * u[k-1]")
```

```python
def gen_prbs(length, min_run=5, max_run=20):
    prbs = []
    current = np.random.randint(0, 2)
    while len(prbs) < length:
        run_length = np.random.randint(min_run, max_run)
        run_length = min(run_length, length - len(prbs))  # Evitar exceder o comprimento
        prbs.extend([current] * run_length)
        current = 1 - current  # Alterna entre 0 e 1
    return np.array(prbs[:length])

# Gerar Sinal PRBS
u_prbs = gen_prbs(num_samples, min_run=5, max_run=20)

# Inicializar Sinal de Saída
y_prbs = np.zeros(num_samples)

# Aplicar a Equação a Diferenças
for k in range(1, num_samples):
    if k == 1:
        # Para k=1, u[k-1] existe, mas u[k-2] não (assume-se u[-1] = 0)
        y_prbs[k] = a1 * y_prbs[k-1] + b0 * u_prbs[k] + b1 * 0
    else:
        y_prbs[k] = a1 * y_prbs[k-1] + b0 * u_prbs[k] + b1 * u_prbs[k-1]

# Plotar Sinal PRBS e Resposta do Sistema
plt.figure(figsize=(12, 8))

plt.subplot(2, 1, 1)
plt.step(t_d, u_prbs, where='post', label='Entrada PRBS', color='blue')
plt.xlabel('Tempo (s)')
plt.ylabel('Amplitude')
plt.title('Entrada PRBS')
plt.grid(True)

plt.subplot(2, 1, 2)
plt.step(t_d, y_prbs, where='post', label='Saída y[k]', color='green')
plt.xlabel('Tempo (s)')
plt.ylabel('Amplitude')
plt.title('Resposta do Sistema à Entrada PRBS')
plt.grid(True)

plt.tight_layout()
plt.show()
```
